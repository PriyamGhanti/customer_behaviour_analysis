```python
#filling null values by median because here mean purity can be affected by outliers an dalso affect dadta consistency
#and also with each category respectively to ensure one category review rating is not interfered by the another
df['Review Rating']  = df.groupby('Category')['Review Rating'].fillna
```

```python
df['review_rating'] = df.groupby('category')['review_rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'Purchase_Amount'})
```

```python
df.columns
```

```python
# creating a new column for age category
labels = ['Young Adult','Adult','Middle-Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels =labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping = {'Fortnightly':14,
                        'Weekly':7,
                        'Monthly':30,
                        'Quarterly':90,
                        'Bi-Weekly':14,
                        'Annually':365,
                         'Every 3 Months':90
                       }
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
